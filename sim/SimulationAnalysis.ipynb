```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```python
def convert_to_int(string):
    num = int(string.strip(),2)
    val = num & 0x7FF
    if(num > 0x7FF):
        val = -1*(0x800 - val)
    return val
```

```python
def csv_to_data(filename):
    sample_in = []
    filter_A = []
    filter_B = []
    with open(filename, "r") as infile:
        for line in infile:
            line = line.split(",")
            for idx in range(8+8+8):
                try:
                    if(idx < 8):
                        sample_in.append(convert_to_int(line[idx]))
                    elif(idx < 16):
                        filter_A.append(convert_to_int(line[idx]))
                    else:
                        filter_B.append(convert_to_int(line[idx]))
                except:
                    break
    return np.array(sample_in), np.array(filter_A), np.array(filter_B)
```

# Direct Comparison

```python
for i in range(8):
    sample_in, filter_A, filter_B = csv_to_data("04_12_2024_V2/simulation_output_in_%d.txt"%i)
    clocks = np.arange(len(sample_in))
    filter_A_start = np.min(np.argwhere(filter_A>0))
    filter_B_start = np.min(np.argwhere(filter_B>0))
    filter_start = np.min(np.argwhere(sample_in>0))
    plt.legend()
    samples = 55
    rescale = 2
    # plt.plot(clocks[filter_A_start-50:filter_A_start-50+samples], sample_in[filter_A_start-50:filter_A_start-50+samples], linestyle="-", marker=".", label="Input")
    plt.plot(clocks[filter_A_start-10:filter_A_start-10+samples]-filter_start, filter_A[filter_A_start-10:filter_A_start-10+samples], linestyle="-", marker=".", label="Filter A (Custom)")
    plt.plot(clocks[filter_A_start-10:filter_A_start-10+samples]-filter_start, rescale*filter_B[filter_B_start-10:filter_B_start-10+samples], linestyle=":", marker=".", label="Filter B (IP*2)")
    plt.legend()
    plt.xlabel("Clocks")
    plt.title("Impulse response on input %d"%i)
    plt.savefig("plots/impulse_%d.png"%i)
    # plt.xlim(845,875)
    plt.show()
    plt.clf()
```
